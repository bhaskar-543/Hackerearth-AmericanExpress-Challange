```
li=[i for i in k if len(list(data[i].unique()))>2]
```

```
def impute(cols,x,y):
  for i in range(len(cols)):
    cols[i]=(cols[i]-x)/y
  
  return cols
    
```

```
for i in li:
  data[i]=impute(data[i],data[i].mean(),data[i].std())
  
```
